```python
row = 4
col = 6
image = np.array([[random.random() for i in range(col)] for j in range(row)])
```

```python
coord_list = []

def find_coords(x, y):
    """
    Finds 8-neighbors of a given coordinate
    """
    x_range = range(x-1, x+2)
    for i in x_range:
        for j in range(y-1, y+2):
            coord_list.append([i, j])

find_coords(2,1)
coord_list
```

```python
vals = []
def find_vals(image, coords):
    """
    Finds values in image's neighbors given list of neighbor's coordinates
    """
    for i in coords:
        try:
            vals.append(image[i[0]][i[1]])
        except IndexError:
            print("error on image edge - ignoring for now")
            continue
find_vals(image, coord_list)
```

```python
np.mean(vals)
```

```python
image.shape[0] # 4 row, 6 col
```

```python
rows = image.shape[0]
cols = image.shape[1]

new_image = [ [0]*6 for i in range(4)]
for y in range(cols):
    for x in range(rows):
        # create empty lists to store our coordinates and values to average later
        coord_list = []
        vals = []

        # find coords neighbor
        find_coords(x, y)

        # find all values from these coords
        find_vals(image, coord_list)

        # store this value in new image
        #print(x, y)
        new_image[x][y] = np.mean(vals)
        
```
